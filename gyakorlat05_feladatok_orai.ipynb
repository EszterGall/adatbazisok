1. Listázzuk az egyes csoportokat, és a csoportokba tartozó termékek darabszámát!
2. A lista jelenítse meg a végösszeget is!

````sql
SELECT
    case 
        when listaar>5000 then 'drága'
        when listaar<3000 then 'olcsó'
        else 'közepes'
    end,
    count(termekkod)
from Termek
group by rollup(case
                when listaar>5000 then 'drága'
                when listaar<3000 then 'olcsó'
                else 'közepes'
                end)
````
